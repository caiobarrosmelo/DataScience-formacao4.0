Média de Alunos por Turma da Educação Básica - Ensino Infantil (Creche, Pré-Escola), Ensino Fundamental e Ensino Médio, por Localização e Dependência Administrativa - Brasil, Regiões Geográficas e Unidades da Federação - 2024																									


## Carregando o conjunto de dados

```python
import pandas as pd
atu = pd.read_excel('datasets/ATU_BRASIL_REGIOES_UFS_2024.xlsx', skiprows=8)
atu
```

```python
atu.info()
```

## Preparando os dados para análise

**Obs.:** *Todas as colunas estão sendo identificadas como 'Object'.*

Antes de resolver esse problema, vamos selecionar as colunas 'Unidades Geográfica', 'Localização', 'Dependência Administrativa', 'Total Infantil', 'Total Fundamental' e 'Total Médio para efetuar a análise'.

```python
atu = atu.iloc[:,[1,2,3,4,7,20]]

colunas = ['UG', 'LOC', 'DA', 'TI', 'TF', 'TM']
```

Dicionário:
- ug: Unidade Geográfica
- loc: localicação
- da: Dependência A
- ti: Total Ensino Infantil
- tf: Total Ensino Fundamental
- tm: Total Ensino Médio

```python
colunas = ['ug', 'loc', 'da', 'ti', 'tf', 'tm']
atu.columns = colunas
atu
```

```python
atu.info()
```

Vamos considerar que os valores ausentes indicam que não existem escolas com aquele nível de ensino na região.

```python
df = atu.replace('--', 0)
```

```python
df.info()
```

```python
df.groupby('ug')['ti'].median()
```

```python
df.groupby('ug')['ti'].mean()
```

```python
df.groupby('ug')[['ti', 'tf', 'tm']].median().round(2)
```

```python
df.groupby('loc')[['ti', 'tf', 'tm']].median().round(2)
```

```python
df.groupby(['ug', 'loc'])[['ti', 'tf', 'tm']].median().round(2)
```

```python
pd.pivot_table(df, values='tf', index='ug', columns='da', aggfunc='median').round(2)
```

## Visualização dos dados

**Gráfico de barras (barra horizontal)**

Comparar médias por Tipo de Escola(DA) ou localização(LOC)

```python
import matplotlib.pyplot as plt
 
df.groupby("da")[["ti", "tf", "tm"]].mean().plot(kind="barh", figsize=(10, 5))
plt.title("Média de alunos por Dependência Administrativa")
plt.xlabel("Quantidade média de matrículas")
plt.ylabel("Tipo de administração")
plt.tight_layout()
plt.show()
```

**Gráfico de Colunas Agrupadas**

Comparar múltiplas variáveis por categoria

```python
df.groupby("loc")[["ti", "tf", "tm"]].mean().plot(kind="bar", figsize=(8, 5))
plt.title("Média de matrículas por Localização")
plt.ylabel("Média de alunos")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

**Gráfico de pizza**

Distribuição percentual por Dependência Administrativa

```python
df.groupby("da")["ti"].mean().plot.pie(autopct="%1.1f%%", figsize=(6, 6), ylabel="")
plt.title("Distribuição da Dependência Administrativa para o total Infantil")
plt.show()
```
